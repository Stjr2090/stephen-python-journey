WHILE LOOP

```python
i=0
while i<3:
  print ("meow")
  i+=1
```

```python
while True:
  n= int(input("what's n?"))
  if n>0:
    break
for _ in range(n):
  print ("meow")
```

```python
def main():
  number= get_number()
  meow(number)

def get_number():
  while True:
    n= int(input("what's n? "))
    if n>0:
      break
  return n

def meow (n):
  for _ in range(n):
    print ("meow")

main()
```
